```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#read the data
df = pd.read_csv('movies.csv')
df.head()
```

```python
#find if there are any missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
print(df['budget'].isna().sum())  # Check for NaN
print(np.isinf(df['budget']).sum())  # Check for inf
```

```python
for col in ['budget', 'gross', 'votes']:
    df[col] = df[col].fillna(0)  
    df[col] = df[col].replace([np.inf, -np.inf], 0)  
    df[col] = df[col].astype('int64') 
```

```python
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df['year_correct'] = df['released'].astype(str).str[:35]
```

```python
df.sort_values(by=['gross'], inplace=False , ascending=False)
```

```python
pd.set_option('display.min_rows', None)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
plt.scatter(df['budget'],df['gross'],color='green')

plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
sns.regplot(x='gross', y='budget', data= df, scatter_kws={'color': 'red'}, line_kws={'color' :'blue'})
plt.show()
```

```python
numeric_cols = ['budget', 'gross', 'votes']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

correlation_matrix = df.select_dtypes(include=np.number).corr()
print(correlation_matrix)
```

```python
correlation_matrix = df.select_dtypes(include=np.number).corr()
plt.title('Correlation matric for Numeric features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
